```
import sys
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
#Function Definitions

#function with a local min
def f(x,y):
  return x**2 + y**2

#function with a local max
def g(x,y):
  return -(x**2 + y**2)

#function with neither a min or max, but a point where the gradient is 0
def h(x,y):
  return x**4 - y**4
```

```
#local min function
def localMin(X,Y,Z):
  minX = sys.maxint
  minY = sys.maxint
  minZ = sys.maxint
    
  for i in range(len(Z)):
    for j in range(len(Z[i])):
      if minZ > Z[i][j]:
        minX = X[i][j]
        minY = Y[i][j]
        minZ = Z[i][j]
    
  return minX,minY,minZ
```

```
#local max function
def localMax(X,Y,Z):
  maxX = -sys.maxint-1
  maxY = -sys.maxint-1
  maxZ = -sys.maxint-1
    
  for i in range(len(Z)):
    for j in range(len(Z[i])):
      if maxZ < Z[i][j]:
        maxX = X[i][j]
        maxY = Y[i][j]
        maxZ = Z[i][j]
    
  return maxX,maxY,maxZ
```

```
#f(x,y)
x = np.linspace(-10, 10, 7)
y = np.linspace(-10, 10, 7)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#finding the min
minX, minY, minZ = localMin(X,Y,Z)

#finding the gradient
fArray = np.concatenate((X, Y, Z))
fGrad = np.gradient(fArray)

#plotting as a 2D contour
plt.title("2D contour for function f")
plt.contour(X, Y, Z, colors='black')
plt.show()

#plotting as a 3D contour
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("3D contour for function f")
ax.plot_surface(X, Y, Z)
ax.contour(X, Y, Z)
plt.show()

#plotting the minimum
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Minimum for function f (see dot)")
ax.plot_wireframe(X, Y, Z)
ax.contour(X, Y, Z)
ax.scatter(minX,minY,minZ,s=500)
plt.show()


#plotting the gradient as vector field
plt.figure()
plt.title("vector field for the gradient of function f")
plt.quiver(X,Y,fGrad[1],fGrad[0])
plt.show()
```

```
#g(x,y)
x = np.linspace(-10, 10, 7)
y = np.linspace(-10, 10, 7)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)

#finding the max
minX, minY, minZ = localMax(X,Y,Z)

#finding the gradient
fArray = np.concatenate((X, Y, Z))
fGrad = np.gradient(fArray)

#plotting as a 2D contour
plt.title("2D contour for function g")
plt.contour(X, Y, Z, colors='black')
plt.show()

#plotting as a 3D contour
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("3D contour for function g")
ax.plot_surface(X, Y, Z)
ax.contour(X, Y, Z)
plt.show()

#plotting the minimum
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Maximum for function g (see dot)")
ax.plot_wireframe(X, Y, Z)
ax.contour(X, Y, Z)
ax.scatter(minX,minY,minZ,s=500)
plt.show()


#plotting the gradient as vector field
plt.figure()
plt.title("vector field for the gradient of function f")
plt.quiver(X,Y,fGrad[1],fGrad[0])
plt.show()
```
